```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y=train["Survived"]
f=["Pclass","Sex","SibSp","Parch"]
x=pd.get_dummies(train[f])
x_test=pd.get_dummies(test[f])
```

```python
mlr=RandomForestClassifier(n_estimators=200,max_depth=8,random_state=1)
mlr.fit(x,y)
predict=mlr.predict(x_test)
```

```python
output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predict})
output.to_csv('submission.csv', index=False)
```
